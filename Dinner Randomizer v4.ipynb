### create meal list based on preferences

```python
list_col = df_meals.columns.tolist()
#list_col
df_multi = pd.DataFrame(columns=list_col)
df_multi
```

```python
# loop thru df
for index, row in df_meals.iterrows():
     for i in range(row['Rating']):
            #print(index, row['MealName'], row['Rating'])
            df_multi = df_multi.append(
                {
                    'MealID': row['MealID'], 
                    'MealName': row['MealName'], 
                    'Carb': row['Carb'],
                    'Protein': row['Protein'],
                    'Veg': row['Veg'],
                    'Ease': row['Ease'],
                    'Rating': row['Rating'],
                    'MealGroup': row['MealGroup']
                },
                ignore_index=True
            )
            
df_multi.head(10)
```
